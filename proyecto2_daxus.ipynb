```python
data = yfinance.Ticker("AAPL").history("6mo") #Información de los últimos 6 meses de Apple
data
```

```python
data.Close # Muestra datos de la columna "Close"
```

## Hacer el anáisis solicitado por el gestor
* Cotización Máxima
* Cotización Mínima
* Cotización Media de Cierre de la acción

```python
cierre # Información de cierre
```

```python
maxima = round(cierre.max(), 2) # Cotización máxima
minima = round(cierre.min(), 2) # Cotización mínima
valor_medio = round(cierre.mean(), 2) #Cotización media
```

```python
print(maxima)
print(minima)
print(valor_medio)
```

## Enviar correo electrónico automáticamente
* Abre una nueva pestaña
* Entrar al correo electrónico
* Haz click en el botón de  redactar
* Escribimos la dirección del destinatario
* Escribimos asunto
* Redactamos correo
* Haz click en "enviar"

```python
!pip install pyautogui
!pip install pyperclip
```

```python
import pyautogui
import pyperclip
import webbrowser
import time #tiempo de reacción
```

```python
webbrowser.open(https://mail.google.com/mail/u/0/?pli=1#inbox)
destinatario = "micorreo@gmail.com"
asunto = "Análisis de acciones"
cuerpo = f"""
Buenas tardes Zai,

Acá te envío el análisis de las acciones de los últimos 6 meses de Apple
Cotización máxima: USD {maxima}
Cotización mínima: USD {minima}
Valor medio: USD {valor_medio}

¡Estoy pendiente a cualquier observación!
Saludos,
Daniel
""" # Triple comilla para poder hacer saltos de línea

time.sleep(3) # tiempo de 3 segundos
pyautogui.PAUSE = 3 # cantidad de segundos por cada vez que se utilice pyautogui
pyautogui.click(3720,617)
pyperclip.copy(destinatario) #copiar
pyautogui.hotkey("ctrl","v") # pegar
pyautogui.hotkey  ("tab")

pyperclip.copy(asunto) #copiar
pyautogui.hotkey("ctrl","v") # pegar
pyautogui.hotkey  ("tab")

pyperclip.copy(cuerpo) #copiar
pyautogui.hotkey("ctrl","v") # pegar

pyautogui.click(5237,1718)

pyautogui.hotkey("ctrl","w")
```
